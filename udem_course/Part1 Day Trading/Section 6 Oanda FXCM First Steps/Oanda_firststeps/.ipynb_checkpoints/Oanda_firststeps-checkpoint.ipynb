## Streaming high-frequency real-time Data

```python
api.stream_data('EUR_USD', ) # doable when market is open, otherwise retrieves last data and waits
```

## Creating Orders and Executing Trades

```python
import pandas as pd
import tpqoa
```

```python
api = tpqoa.tpqoa("oanda.cfg")
```

```python
api.create_order(instrument = "EUR_USD", units = 1000, sl_distance= 0.1) #open the position
```

```python
api.create_order(instrument = "EUR_USD", units = -1000, sl_distance= 0.1) # close the position
```

```python
api.get_account_summary()
```

```python
api.get_transactions(tid = 4)
```

```python
api.print_transactions()#tid = 5-1)
```

```python
api.create_order(instrument = "EUR_USD", units = -100, sl_distance= 0.1) # close the position
```

```python
api.print_transactions()#tid = 5-1)
```
